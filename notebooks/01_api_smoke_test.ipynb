```python
import requests
import pandas as pd
```

```python
url = "https://www.airportal.go.kr/airport/getFlightArrivalDepartureInfo.do"

payload = {
    "searchContext": "departure",  
    "ymd": "20240104",
    "startTime": "0000",
    "endTime": "2400",
    "loadType": "json",
    "depAirport": "RKSI"
}

headers = {
    "User-Agent": "Mozilla/5.0",
    "Referer": "https://www.airportal.go.kr/airport/aircraftInfo.do",
    "X-Requested-With": "XMLHttpRequest",
    "Content-Type": "application/json;charset=UTF-8"
}

session = requests.Session()
session.get("https://www.airportal.go.kr/airport/aircraftInfo.do")  # 세션 쿠키 획득

res = session.post(url, json=payload, headers=headers)
print(res.status_code)
data = res.json()
```

```python
data = data['content']
```

```python
df = pd.DataFrame(data)
print(df)
```
